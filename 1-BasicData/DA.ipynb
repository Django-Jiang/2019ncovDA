## 1. Read Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#import scipy.io as sio
#import seaborn as sns
```

```python
filename = "2019ncov.csv"
data = pd.read_csv(filename)
```

```python
data
```

### Update CSV Data

```python
def update(data, newval):
    df_new = pd.DataFrame(newval,columns = ['Date','DomesticInfected','Suspected','Dead','Cured'])
    data2 = data.append(df_new, ignore_index=True)
    return data2

data = update(data,[['2/10', 42638, 21675, 1016,3996],['2/11', 44653, 16067, 1113, 4740]])


data.to_csv("2019ncov.csv", index = 0) ## index = 0 drop the index col when saving
```

##  3. Plot the data

```python
data2 = data[['Date','Dead','Cured']] #take certain col of the data

data2.plot()
plt.show()


data.plot(kind= 'bar')
plt.show()

data.plot(kind = 'bar', stacked = 1)
plt.show()

data.plot(kind = 'area')
plt.show()
```
